# Visualization

```python
def display_result(list_result, path_label):
  for result in list_result:
    test_name = os.path.splitext(result.split('/')[-1])[0]
    with open(path_label+'/'+test_name+'.json', "r") as f:
      dict_label = json.load(f)
      for obj in dict_label["annotations"]:
        class_name = obj["class"]
        status = "normal"
        if obj["attributes"]["status"] != "normal":
          status = obj["attributes"]["status"]
      display(Img(filename=result))
      print(f"[{class_name}] {status}\n")
```

```python
PATH_LABEL = "/content/drive/MyDrive/canAiry/data/02_2nd-dataset/04-1_TR-bearing/01_label/test"
PATH_RESULT = PATH_PROJECT+'/'+CLASS_NAME
results = glob(PATH_RESULT + '/*.jpg')
```

```python
display_result(results, PATH_LABEL)
```

```python
def generate_gif():
  img_list = [PATH_PROJECT+'/'+CLASS_NAME+'/'+x for x in os.listdir(PATH_PROJECT+'/'+CLASS_NAME)[1:]]
  images = [Image.open(x) for x in img_list]
  # loop Î∞òÎ≥µ ÌöüÏàò
  # duration ÌîÑÎ†àÏûÑ Ï†ÑÌôò ÏÜçÎèÑ (500 = 0.5Ï¥à)
  img = images[0]
  img.save(PATH_PROJECT+'/GIF/'+CLASS_NAME+'.gif', save_all=True, append_images=images[1:],loop=1, duration=100)
  return
```
