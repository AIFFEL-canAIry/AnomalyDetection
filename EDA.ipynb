```python
import glob
import json
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def fuck_img(fuck_img_path, fuck_json_path):
  fig, ax = plt.subplots()

  files = glob.glob(fuck_img_path+".csv")
  df = pd.read_csv(files[0])[4:].reset_index(drop=True)
  image = np.zeros((512, 640))
  for idx, row in df.iterrows():
    image[idx, :] = np.array([float(x) for x in row[0].split(';')[:-1]])
    ax.imshow(image, 'jet')

  fuck_json = fuck_json_path+".json"

  with open(fuck_json, "r") as a:
    dict_json = json.load(a)   
    for i, obj in enumerate(dict_json["annotations"]):
      bbox = list(dict_json["annotations"][i]["data"].values())
      x, y, w, h = bbox
      rect = patches.Rectangle((x, y), w, h, linewidth=1, edgecolor='r', facecolor='none')
 
      ax.add_patch(rect)
```
