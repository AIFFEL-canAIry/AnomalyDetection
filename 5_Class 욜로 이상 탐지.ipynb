##GIF 이미지 만들기

```python
import os
from PIL import Image
from IPython.display import Image as Img
from IPython.display import display
```

```python
def generate_gif(path):
    img_list = os.listdir(path)[:121]
    img_list = [path + '/' + x for x in img_list]
    images = [Image.open(x) for x in img_list]
    
    im = images[0]
    im.save('/content/drive/MyDrive/MOF.gif', save_all=True, append_images=images[1:],loop=0xff, duration=100)
    # loop 반복 횟수
    # duration 프레임 전환 속도 (500 = 0.5초)
    return Img(url='out2.gif')
```

```python
generate_gif('drive/MyDrive/pic/yolov5/runs/detect/exp9')
```
