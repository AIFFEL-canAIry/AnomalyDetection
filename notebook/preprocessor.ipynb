```python
import glob
from PIL import Image
```

```python
target_size=(256,256)
```

```python
PATH_TRAIN_IMAGE = "/content/drive/MyDrive/data/sample/train/images/"
PATH_VALID_IMAGE = "/content/drive/MyDrive/data/sample/valid/images/"
PATH_TEST_IMAGE = "/content/drive/MyDrive/data/sample/test/images/"
```

```python
list_file_image = glob.glob(PATH_NORMAL+"*.jpg")
```

```python
for file_image in list_file_image:
  old_img=Image.open(file_image)
  new_img=old_img.resize(target_size,Image.ANTIALIAS)
  new_img.save(file_image, "JPEG")
```
