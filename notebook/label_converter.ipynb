```python
import json
import glob
```

```python
PATH_TRAIN_LABEL_JSON = "/content/drive/MyDrive/data/sample/train/labels/json/"
PATH_TRAIN_LABEL = "/content/drive/MyDrive/data/sample/train/labels/"
PATH_VALID_LABEL_JSON = "/content/drive/MyDrive/data/sample/valid/labels/json/"
PATH_VALID_LABEL = "/content/drive/MyDrive/data/sample/valid/labels/"
```

```python
def coco_to_yolo(bbox, img_size):
    return [((2*bbox[0] + bbox[2])/(2*img_size[0])) , ((2*bbox[1] + bbox[3])/(2*img_size[1])), bbox[2]/img_size[0], bbox[3]/img_size[1]]
```

# Train Dataset

```python
list_file_json = glob.glob(PATH_TRAIN_LABEL_JSON+"*.json")
```

```python
for file_json in list_file_json:
  with open(file_json) as r:
    dict_json = json.load(r)

    class_id = int(dict_json["metadata"]["facility_id"])    
    bbox = list(dict_json["annotations"][0]["data"].values())
    img_size = [dict_json["metadata"]["width"], dict_json["metadata"]["height"]]
    
    X, Y, W, H = coco_to_yolo(bbox, img_size)
    label = f"{class_id-1} {X} {Y} {W} {H}" 

    file_name = dict_json["image"]["filename"]
    file_name = file_name.replace(".jpg", "").strip()

    with open(PATH_TRAIN_LABEL+file_name+".txt", "w") as w:
      w.write(label)
```
