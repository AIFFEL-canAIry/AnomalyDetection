```python
import os
from glob import glob
from tqdm import tqdm
import json
from PIL import Image
from IPython.display import Image as Img
from IPython.display import display
```

```python
def display_result(list_result, path_label):
  for result in list_result:
    test_name = os.path.splitext(result.split('/')[-1])[0]
    with open(path_label+'/'+test_name+'.json', "r") as f:
      dict_label = json.load(f)
      for obj in dict_label["annotations"]:
        class_name = obj["class"]
        status = "normal"
        if obj["attributes"]["status"] != "normal":
          status = obj["attributes"]["status"]
      display(Img(filename=result))
      print(f"[{class_name}] {status}\n")
```

```python
PATH_LABEL_CLASS_0 = "/content/drive/MyDrive/canAiry/data/02_2nd-dataset/03-1_IC-boiler/01_label/test"
PATH_LABEL_CLASS_2 = "/content/drive/MyDrive/canAiry/data/02_2nd-dataset/03-2_IC-pump/01_label/test"

PATH_RESULT_CLASS_0 = '/content/drive/MyDrive/object_detection/results/detect/YOLOv7/IC-boiler'
PATH_RESULT_CLASS_2 = '/content/drive/MyDrive/object_detection/results/detect/YOLOv7/IC-pump'

result_class_0 = glob(PATH_RESULT_CLASS_0 + '/*.jpg')
result_class_2 = glob(PATH_RESULT_CLASS_2 + '/*.jpg')
```

```python
display_result(result_class_0, PATH_LABEL_CLASS_0)
```

```python
display_result(result_class_2, PATH_LABEL_CLASS_2)
```

# GIF 생성

```python
def generate_gif(path_result, class_name):
  img_list = os.listdir(path_result)
  img_list = [path_result + '/' + x for x in img_list]
  images = [Image.open(x) for x in img_list]
  
  im = images[0]
  im.save(path_result+'/'+class_name+'.gif', save_all=True, append_images=images[1:],loop=0xff, duration=500)
  # loop 반복 횟수
  # duration 프레임 전환 속도 (500 = 0.5초)
  return
```

```python
generate_gif(PATH_RESULT_CLASS_0, 'SS-mof')
generate_gif(PATH_RESULT_CLASS_2, 'SS-PT')
```
